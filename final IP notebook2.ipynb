```python
import pandas as pd
df=pd.read_csv('uncleaned_245k_tweets.csv')
df
```

```python
cleandf = df[['content','date']]
cleandf.dropna(subset=['content','date'], inplace=True)
cleandf
```

```python
cleandf['date'] = pd.to_datetime(cleandf['date'], format='%Y-%m-%d')
cleandf
```

```python
cleandf['date'] = cleandf['date'].astype(str)
date_series = cleandf['date'].str[:10]
cleandf['date'] = date_series
cleandf
```

```python
import pandas as pd
from tqdm import tqdm
import numpy as np

chunks = np.array_split(cleandf, 100)
```

```python
from textblob import TextBlob
def textblob_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
# textblob analysis

for chunk in tqdm(chunks):
    # Apply your code to the current chunk
    chunk['textblob_sentiment'] = chunk['content'].apply(textblob_sentiment)
```

```python
chunk1df = pd.concat(chunks, ignore_index=True)
chunk1df
```

```python
sentiments = cleandf.copy()
sentiments['textblob_sentiment'] = chunk1df['textblob_sentiment']
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
def vader_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    return analyzer.polarity_scores(text)
```

```python
sample = cleandf.sample(n=10)
for i in tqdm(sample):
    # Apply your code to the current chunk
    sample['vader_sentiment'] = sample['content'].apply(vader_sentiment)
sample
```

```python
chunks = np.array_split(cleandf, 100)
for chunk in tqdm(chunks):
    # Apply your code to the current chunk
    chunk['vader_sentiment'] = chunk['content'].apply(vader_sentiment)
```

```python
chunk2df = pd.concat(chunks, ignore_index=True)
chunk2df
```

```python
sentiments['vader_sentiment'] = chunk2df['vader_sentiment']
sentiments
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
text = cleandf['content'].tolist()
```

```python
from transformers import pipeline
pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
final = pd.read_csv('sentimentdf.csv')
final
```

```python
final = final.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
final
```

```python
import pandas as pd

# Assuming your date column is named 'date' and it is in datetime format
start_date = pd.to_datetime('2020-02-28')
end_date = pd.to_datetime('2020-07-30')

# Filter the dataframe based on the date range
filtered_df = final[(final['date'] >= start_date) & (final['date'] <= end_date)]
```

```python
filtered_df
```

```python
for i in tqdm(filtered_df):
    sentiments_df = pd.DataFrame(filtered_df['vader_sentiment'].tolist(), index=filtered_df.index)
    sentiments_df[0] = sentiments_df[0].apply(eval)  # Convert string to dictionary
    sentiments_df['neg'] = sentiments_df[0].apply(lambda x: x['neg'])
    sentiments_df['neu'] = sentiments_df[0].apply(lambda x: x['neu'])
    sentiments_df['pos'] = sentiments_df[0].apply(lambda x: x['pos'])
    sentiments_df['comp'] = sentiments_df[0].apply(lambda x: x['compound'])
sentiments = pd.concat([filtered_df, sentiments_df], axis=1)
```

```python
sentiments
```

```python
sentiments = sentiments.drop(columns=['vader_sentiment', 0])
sentiments
```

```python
for i in tqdm(filtered_df):
    sentiments_df = pd.DataFrame(filtered_df['bert_sentiment'].apply(eval).tolist(), index=filtered_df.index)
    sentiments_df['label'] = sentiments_df[0].apply(lambda x: x['label'])
    sentiments_df['score'] = sentiments_df[0].apply(lambda x: x['score'])
sentiments2 = pd.concat([filtered_df, sentiments_df], axis=1)
```

```python
sentiments2
sentiments2 = sentiments2.drop(columns=['bert_sentiment', 0])
sentiments2
```

```python
combinedf = sentiments.copy()
combinedf = combinedf.drop(columns=['bert_sentiment'])
combinedf['bert_label'] = sentiments2['label']
combinedf['bert_score'] = sentiments2['score']
combinedf
```

```python
# Create a mapping dictionary for column renaming
column_mapping = {
    'neg': 'vader_neg',
    'neu': 'vader_neu',
    'pos': 'vader_pos',
    'comp': 'vader_comp'
}

# Use the 'rename()' function to rename the columns
combinedf = combinedf.rename(columns=column_mapping)
combinedf
```

```python
grouped_df = combinedf.groupby('date')['textblob_sentiment'].mean()
tb_df = pd.DataFrame({'date': grouped_df.index, 'textblob_sentiment': grouped_df.values})
tb_df
```

```python
# color designations from now on RGB - red textblob, green vader, blue bert
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

tb_df['date'] = pd.to_datetime(tb_df['date'])
dates = tb_df['date']
scores = tb_df['textblob_sentiment']

fig, ax = plt.subplots()
ax.plot(dates, abs(scores), color='red')

# Set the x-axis tick locator to show only the beginning of each month
ax.xaxis.set_major_locator(mdates.MonthLocator())

# Set the date format of the x-axis tick labels
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Set the title and axis labels
ax.set_title('Textblob Sentiment Score vs Date')
ax.set_xlabel('Date (Month of 2020)')
ax.set_ylabel('Textblob Sentiment Score (absolute value)')

plt.show()
```

```python
max_row = abs(tb_df['textblob_sentiment']).idxmax()
date_with_highest_score2 = tb_df.loc[max_row, 'date']
highest_score2 = tb_df.loc[max_row, 'textblob_sentiment']
print(date_with_highest_score2, highest_score2)
```
